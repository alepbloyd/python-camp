First we import the `json` library.

````python
import json
````

Next we open the file for writing, and use the `json.dump()` method to write our list of dictionaries to the file. 

In the code below, I've used the variable `my_team` to refer to the list of dictionaries created above. If you named your list something different, use that name in place of `my_team`.

Note that we're saving the data to a file called `team-database.json`. The initial `./` indicates that this file will live in the same directory as this notebook.
